# 把mat文件合并成csv(China)

```python
import os 
import pandas as pd 
if not os.path.exists('China'):
    os.mkdir('China')
import scipy.io as sio
def date_mat_in(date):
    path='./百度迁徙/清洗过后的数据/'+date+'_in.mat'
    df= sio.loadmat(path)
    path1='./百度迁徙/keyset/keyset1.mat'
    df1 = sio.loadmat(path1)

    xn,yn=df['data_in'].shape
    for i in range(xn):
        country=df1['keyset'][2][i][0]
        city=df1['keyset'][0][i][0]
        country_os='./China/'+country
        if not os.path.exists(country_os):
            os.mkdir(country_os)
        path2='./China/'+country+'/'+city+'_in'+'.csv'
        if not os.path.exists(path2):
            pd.DataFrame(df['data_in'][i]).to_csv(path2,index=0)
            df2=pd.read_csv(path2)
            df2.columns=['n']
            df2.drop('n',axis=1,inplace=True)
            df2[date]=df['data_in'][i]
            df2.to_csv(path2,index=0)
        else:
            df2=pd.read_csv(path2)
            df2[date]=df['data_in'][i]
            df2.to_csv(path2,index=0)
def date_mat_out(date):
    path='./百度迁徙/清洗过后的数据/'+date+'_out.mat'
    df= sio.loadmat(path)
    path1='./百度迁徙/keyset/keyset1.mat'
    df1 = sio.loadmat(path1)

    xn,yn=df['data_out'].shape
    for i in range(xn):
        country=df1['keyset'][2][i][0]
        city=df1['keyset'][0][i][0]
        country_os='./China/'+country
        if not os.path.exists(country_os):
            os.mkdir(country_os)
        path2='./China/'+country+'/'+city+'_out'+'.csv'
        if not os.path.exists(path2):
            pd.DataFrame(df['data_out'][i]).to_csv(path2,index=0)
            df2=pd.read_csv(path2)
            df2.columns=['n']
            df2.drop('n',axis=1,inplace=True)
            df2[date]=df['data_out'][i]
            df2.to_csv(path2,index=0)
        else:
            df2=pd.read_csv(path2)
            df2[date]=df['data_out'][i]
            df2.to_csv(path2,index=0)
```

```python
#提取日期date列表
path='./百度迁徙/清洗过后的数据/'
date_num=os.listdir(path)
date=[]
import re 
for i in date_num:
    x=re.findall('(.*?)_out.mat',i)
    if x!=[]:
        date.append(x[0])
#按地区把日期合并在一起
for i in date:
    date_mat_in(i)
    date_mat_out(i)
    print('clean %s'%i)
    
```

```python
import os
import pandas as pd 
for x in os.listdir('./China_new'):
    print(x)
    for y in os.listdir('./China_new/'+x):
        path='./China_new/'+x+'/'+y
        df=pd.read_csv(path,encoding='gbk',index_col=0)
        all_in=[]  #全部迁入的城市放入一个列表
        all_out=[] #全部迁出的城市放入一个列表
        for i in df.columns.tolist():
            if i.split('_')[1]=='in':
                all_in.append(i)
            if i.split('_')[1]=='out':
                all_out.append(i)
        df['all_in']=df[all_in].sum(axis=1).tolist()  #增加一个总迁入特征
        df['all_out']=df[all_out].sum(axis=1).tolist()  #增加一个总迁出特征
        df['净迁入强度']=df.all_in-df.all_out
        df['净迁出强度']=df.all_out-df.all_in
        df['总迁入_mean']=df[all_in].mean(axis=1).tolist()
        df['总迁出_mean']=df[all_out].mean(axis=1).tolist()
        df['总迁入_var']=df[all_in].var(axis=1).tolist()
        df['总迁出_var']=df[all_out].var(axis=1).tolist()
        df['总迁入_std']=df[all_in].std(axis=1).tolist()
        df['总迁出_std']=df[all_out].std(axis=1).tolist()
        df['总迁入_median']=df[all_in].median(axis=1).tolist()
        df['总迁出_median']=df[all_out].median(axis=1).tolist()
        df['总迁入_max']=df[all_in].max(axis=1).tolist()
        df['总迁出_max']=df[all_out].max(axis=1).tolist()
        df['总迁入_min']=df[all_in].min(axis=1).tolist()
        df['总迁出_min']=df[all_out].min(axis=1).tolist()
        df['总迁入_quantile0.25']=df[all_in].quantile(0.25,axis=1).tolist()  #上四分位数
        df['总迁出_quantile0.25']=df[all_out].quantile(0.25,axis=1).tolist()
        df['总迁入_quantile0.75']=df[all_in].quantile(0.75,axis=1).tolist()  #下四分位数
        df['总迁出_quantile0.75']=df[all_out].quantile(0.75,axis=1).tolist()
        df['总迁入_skew']=df[all_in].skew(axis=1).tolist()  #偏度 偏度（skewness），是统计数据分布偏斜方向和程度的度量，是统计数据分布非对称程度的数字特征。偏度(Skewness)亦称偏态、偏态系数。
        df['总迁出_skew']=df[all_out].skew(axis=1).tolist()
        df['总迁入_kurt']=df[all_in].kurt(axis=1).tolist()  #峰度 峰度（peakedness；kurtosis）又称峰态系数。表征概率密度分布曲线在平均值处峰值高低的特征数。直观看来，峰度反映了峰部的尖度。
        df['总迁出_kurt']=df[all_out].kurt(axis=1).tolist()
        df.to_csv(path,encoding='gbk')
```

# 清洗成用第一周7d预测第二周最后的一天(China_new_week2)

```python
import os
import pandas as pd 
if not os.path.exists('China_new_week2'):
    os.mkdir('China_new_week2')
for x in os.listdir('./China_new2'):
    if not os.path.exists('./China_new_week2/'+x):
            os.mkdir('./China_new_week2/'+x)
    for y in os.listdir('./China_new2/'+x):
        path1='./China_new_week2/'+x+'/'+y
        index=[]
        all_in=[]
        all_out=[]
        path='./China_new2/'+x+'/'+y
        print(path)
        df=pd.read_csv(path,encoding='gbk',index_col=0)
        for i in range(len(df)-14+1):
            df1=pd.DataFrame(df.iloc[i:(7+i),:].values.reshape(1,-1))
            index.append(df.iloc[i:(7+i),:].index[-1])
            all_in.append(df.iloc[13+i:14+i,:]['all_in'].values[0])
            all_out.append(df.iloc[13+i:14+i,:]['all_out'].values[0])
            df1.to_csv(path1,encoding='gbk',mode='a',header=0,index=0) 
        d=pd.DataFrame(df.columns.values)
        d1=d.apply(lambda x:x+'_d1')
        d1['0']=d.apply(lambda x:x+'_d1')
        d1['1']=d.apply(lambda x:x+'_d2')
        d1['2']=d.apply(lambda x:x+'_d3')
        d1['3']=d.apply(lambda x:x+'_d4')
        d1['4']=d.apply(lambda x:x+'_d5')
        d1['5']=d.apply(lambda x:x+'_d6')
        d1['6']=d.apply(lambda x:x+'_d7')
        a=[d1['0'],d1['1'],d1['2'],d1['3'],d1['4'],d1['5'],d1['6']]
        columns=pd.concat(a)
        df1=pd.read_csv(path1,encoding='gbk',header=None)
        df1.columns=columns
        df1.index=index
        all_in_week=[]
        all_out_week=[]
        all_cityinner_week=[]
        for i in df1.columns:
            try:
                if i.split('_')[1]=='in':
                    if i.split('_')[0]=='all':
                        pass
                    else:
                        all_in_week.append(i)
                if i.split('_')[1]=='out':
                    if i.split('_')[0]=='all':
                        pass
                    else:
                        all_out_week.append(i)
                if i.split('_')[0]=='城内活动强度':
                    all_cityinner_week.append(i)
            except:
                continue
        df1['all_in']=all_in
        df1['all_out']=all_out
        df1['all_in_week']=df1[all_in_week].sum(axis=1).tolist()  #增加一个总迁入特征
        df1['all_out_week']=df1[all_out_week].sum(axis=1).tolist()  #增加一个总迁出特征
        df1['净迁入强度_week']=df1.all_in_week-df1.all_out_week
        df1['净迁出强度_week']=df1.all_out_week-df1.all_in_week
        df1['总迁入_mean_week']=df1[all_in_week].mean(axis=1).tolist()
        df1['总迁出_mean_week']=df1[all_out_week].mean(axis=1).tolist()
        df1['总迁入_var_week']=df1[all_in_week].var(axis=1).tolist()
        df1['总迁出_var_week']=df1[all_out_week].var(axis=1).tolist()
        df1['总迁入_std_week']=df1[all_in_week].std(axis=1).tolist()
        df1['总迁出_std_week']=df1[all_out_week].std(axis=1).tolist()
        df1['总迁入_median_week']=df1[all_in_week].median(axis=1).tolist()
        df1['总迁出_median_week']=df1[all_out_week].median(axis=1).tolist()
        df1['总迁入_max_week']=df1[all_in_week].max(axis=1).tolist()
        df1['总迁出_max_week']=df1[all_out_week].max(axis=1).tolist()
        df1['总迁入_min_week']=df1[all_in_week].min(axis=1).tolist()
        df1['总迁出_min_week']=df1[all_out_week].min(axis=1).tolist()
        df1['总迁入_quantile0.25_week']=df1[all_in_week].quantile(0.25,axis=1).tolist()  #上四分位数
        df1['总迁出_quantile0.25_week']=df1[all_out_week].quantile(0.25,axis=1).tolist()
        df1['总迁入_quantile0.75_week']=df1[all_in_week].quantile(0.75,axis=1).tolist()  #下四分位数
        df1['总迁出_quantile0.75_week']=df1[all_out_week].quantile(0.75,axis=1).tolist()
        df1['总迁入_skew_week']=df1[all_in_week].skew(axis=1).tolist()  #偏度 偏度（skewness），是统计数据分布偏斜方向和程度的度量，是统计数据分布非对称程度的数字特征。偏度(Skewness)亦称偏态、偏态系数。
        df1['总迁出_skew_week']=df1[all_out_week].skew(axis=1).tolist()
        df1['总迁入_kurt_week']=df1[all_in_week].kurt(axis=1).tolist()  #峰度 峰度（peakedness；kurtosis）又称峰态系数。表征概率密度分布曲线在平均值处峰值高低的特征数。直观看来，峰度反映了峰部的尖度。
        df1['总迁出_kurt_week']=df1[all_out_week].kurt(axis=1).tolist()
        df1['城内活动强度_week']=df1[all_cityinner_week].sum(axis=1).tolist() 
        df1.to_csv(path1,encoding='gbk')
#         break
#     break

        
```

# 把全部城市合并在一个数据集上(Dataset.csv)

```python
import os
import pandas as pd
def city_all():
    num = 0
    for x in os.listdir('./China_new_week2'):
        for y in os.listdir('./China_new_week2/'+x):
            path='./China_new_week2/'+x+'/'+y
            name = y[:-4]
            df=pd.read_csv(path,encoding='gbk',index_col=0)
            df["city"] = name
            if(num==0):
                data = df
                num = num+1
            else:
                data = pd.concat([data,df])
    data.to_csv("./Dataset.csv")
    return data
df_all=city_all()
df_all
```
